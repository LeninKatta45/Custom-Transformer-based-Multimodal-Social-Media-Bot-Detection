```python
import csv
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

def extract_image_urls(driver, account_name, max_scrolls=10, required_images=4):
    driver.get(f'https://x.com/{account_name}')
    time.sleep(5)
    
    # Check if a "retry" button appears and click it, then go to "media"
    try:
        retry_button = driver.find_element(By.XPATH, "//span[text()='Retry']")
        retry_button.click()
        time.sleep(3)
        
        # After clicking Retry, go to the "Media" section
        media_button = driver.find_element(By.XPATH, "//span[text()='Media']")
        media_button.click()
        time.sleep(5)
        print(f"Media tab clicked for account '{account_name}' after Retry")
    except NoSuchElementException:
        print(f"No retry button for account '{account_name}', continuing the usual process...")

    unique_image_urls = set()
    scroll_count = 0

    while len(unique_image_urls) < required_images and scroll_count < max_scrolls:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        
        images = driver.find_elements(By.CSS_SELECTOR, 'img[src^="https://pbs.twimg.com/media/"]')
        new_image_urls = {img.get_attribute('src') for img in images}
        unique_image_urls.update(new_image_urls)
        
        scroll_count += 1
    
    return list(unique_image_urls)[:required_images]

with open(csv_file, 'a', newline='') as f:
    writer = csv.writer(f)

    if not file_exists:
        writer.writerow(['screen_name', 'posts_url'])
    
    vill = 1    
    # Start from the 90th account onwards
    for c, account_name in enumerate(df['screen_name'][133:], start=133):
        try:
            # Try to extract image URLs
            filtered_urls = extract_image_urls(driver, account_name)
            
            if not filtered_urls:
                print(f"Account '{account_name}' has no images or does not exist.")
                filtered_urls = []  # Or handle accordingly if you want to write something else to CSV

            writer.writerow([account_name, filtered_urls])
            print(f"Iteration-{vill} | URLs Extracted: {filtered_urls}")
        except Exception as e:
            print(f"Error processing account '{account_name}': {e}")
            writer.writerow([account_name, []])  # Handle accounts with errors or exceptions

        vill += 1
        time.sleep(5)

# Close the driver
driver.quit()
```

```python
def extend_array(x):
    x=ast.literal_eval(x)
    print(len(x))
    if (len(x)==0 or len(x)==4):
        return x
    if len(x)==1:
        for i in range(3):
            x.append(x[0])
    if len(x)==2:
        for i in range(1):
            x.append(x[0])
            x.append(x[1])
    if len(x)==3:
        for i in range(1):
            x.append(x[1])
    
    return x
            
```

```python
from datetime import datetime

def calculate_age_in_days(year: int) -> int:
    # Starting date: January 1st of the input year
    start_date = datetime(year, 1, 1)
    
    # End date: December 31st, 2019
    end_date = datetime(2019, 12, 31)
    
    # Calculate the difference in days
    age_in_days = (end_date - start_date).days
    
    return age_in_days
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup Selenium WebDriver (e.g., ChromeDriver)
'''options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode (without opening the browser)
driver = webdriver.Chrome(options=options)'''

def API_scrap_selenium(name_list, count_num):
    fail_lst = []
    user_dfs = {}
    
    for name in name_list:
        user_dfs[name] = pd.DataFrame()
        try:
            # Open Twitter user page
            driver.get(f'https://x.com/{name}')
            time.sleep(5) # Give it time to load
            
            # Scroll to load tweets
            for _ in range(count_num // 5):  # Adjust scrolling based on the number of tweets you want
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)  # Wait for new tweets to load

            # Locate tweet elements (using CSS selectors for tweets)
            tweets = driver.find_elements(By.CSS_SELECTOR, 'article div[data-testid="tweet"]')

            tweet_data = []
            for tweet in tweets[:count_num]:
                try:
                    tweet_text = tweet.find_element(By.CSS_SELECTOR, 'div[lang]').text
                    timestamp = tweet.find_element(By.CSS_SELECTOR, 'time').get_attribute('datetime')
                    tweet_data.append({
                        'user': name,
                        'text': tweet_text,
                        'timestamp': timestamp
                    })
                except Exception as e:
                    print(f"Error scraping tweet for {name}: {str(e)}")
                    continue

            # Add to the DataFrame for the user
            user_dfs[name] = pd.DataFrame(tweet_data)

        except Exception as e:
            print(f"Error scraping user {name}: {str(e)}")
            fail_lst.append(name)
    
    return user_dfs, fail_lst


# Usage
gu_dfs, fail_lst = API_scrap_selenium(gu_list, 10)
ssbots_dfs, ssbots_fail_lst = API_scrap_selenium(ssbots_list, 10)
tsbots_dfs, tsbots_fail_lst = API_scrap_selenium(tsbots_list, 10)
sh_user_dfs, sh_fail_lst = API_scrap_selenium(lu_list, 10)
sh_bot_dfs, sh_bot_fail_lst = API_scrap_selenium(bot_list, 10)

# Close the driver after scraping
driver.quit()
```

```python
import pandas as pd
from datetime import datetime

# Function to extract year from "created_at"
def extract_year(created_at_str):
    return datetime.fromisoformat(created_at_str.replace('+00:00', '+0000')).year

# Prepare the dataset
dataset = []
for entry in data:
    record = {
        'id': entry['id'],
        'created_at_year': extract_year(entry['created_at']),
        'username': entry['username'],
        'profile_image_url': entry['profile_image_url'],
        'profile_banner_url': entry.get('profile_banner_url', entry['profile_image_url']),
        'followers_count': entry['public_metrics']['followers_count'],
        'friends_count': entry['public_metrics']['following_count'],
        'description': entry['description'],
        'tweet_count': entry['public_metrics']['tweet_count'],
    }
    dataset.append(record)

# Convert to a DataFrame
df = pd.DataFrame(dataset)

# Display the dataset
print(df)

# Optionally save the dataset to a CSV file
df.to_csv('twitter_user_dataset.csv', index=False)
```
